```python
# ACDC

from skimage.util import view_as_blocks

def generate_mask(seg_cs):
    print('Shape: ', seg_cs.shape)
    mask = np.ones(seg_cs.shape)
    num_h = 40.
    num_w = 40.
    img_h, img_w = seg_cs.shape
    patch_h = int(img_h/num_h)
    patch_w = int(img_w/num_w)
    print('patch h: ', patch_h)
    print('patch w: ', patch_w)
    patches = view_as_blocks(seg_cs, (patch_h, patch_w))
    
    half_patch_h = int(patch_h/2.)
    half_patch_w = int(patch_w/2.)
    
    center_h = int(patch_h/2.)
    
    for p_1 in patches:
        center_w = int(patch_w/2.)
        for p_2 in p_1:
            if np.mean(p_2)>0.5:
                left_bound = max([0, center_w - half_patch_w])
                right_bound = min([img_w, center_w + half_patch_w])
                upper_bound = max([0, center_h - half_patch_h])
                bottom_bound = min([img_h, center_h + half_patch_h])
                mask[upper_bound:bottom_bound+1, left_bound:right_bound+1] = 0
                
            center_w = center_w + patch_w
        center_h = center_h + patch_h
                
    return mask
                
npy_root = '/home/zheng/Softwares/Experiments/diff/testing_data/cityscapes'
npy_list = sorted(glob.glob(os.path.join(npy_root, '*.npy')))
for npy_path in npy_list:
    seg_logit = np.load(npy_path)
    seg_cs = np.amax(seg_logit, axis=0) # Confidence score
    seg_id = np.argmax(seg_logit, axis=0)
    seg_id = seg_id+1

    inpa_mask = generate_mask((seg_cs<0.9).astype(int))
    
    seg_id[inpa_mask==0] = 0
    
    plt.imshow(seg_cs<0.7)
    plt.show()
    
    inpa_mask_path = os.path.join(npy_root, 'mask.npy')
    np.save(inpa_mask_path, inpa_mask)
    plt.imshow(inpa_mask)
    plt.show()
    
    color_seg = np.zeros((seg_id.shape[0], seg_id.shape[1], 3), dtype=np.uint8)
    for label, color in enumerate(PALETTE_1):
        color_seg[seg_id == label, :] = color
        
    plt.imshow(color_seg)
    plt.show()
```

```python
def projection_bev_matrix(rx=4, rz=-4, ux=100, uz=200):
    """
    rx, rz: grid resolution: number of grid cells per meter in x and z direction
    ux, uz: coordinates of origin of the bev grid
    :return: M - bev_projection_matrix such that M @ np.array([x,z,1]).reshape(3,1) gives the grid cell coordinates

    for rx=4, rz=-4, ux=100, uz=200
    # (x, z) -> (bu, bv)
    # (-25, 0) -> (0, 200)
    # (25, 0) -> (200, 200)
    # (25, 50) -> (200, 0)
    # (-25, 50) -> (0, 0)
    """

    bev_projection_matrix = np.array([[rx, 0, ux],
                                      [0, rz, uz],
                                      [0, 0, 1]])
    return bev_projection_matrix

def project_to_bev(u, z, K):
    """

    :param u: float64 numpy array - (n,) - the u coordinate of points in image coordinate system
    :param z: float64 numpy array - (n,) - the depth of the corresponding point in the camera coordinate system
    :return: 200x200 0/1 matrix representing the BEV grid
    where the grid cells are set to 1, when any point projected on to the BEV plane lies in the cell.

    BEV grid coordinate system is the image coordinate system UV where the origin of the camera coordinate system
    is at (100, 200)
    """
    # project to 3D space
    f, cu = K[0, 0], K[0, 2]
    x = z * (u - cu) / f
    bev_image_points = projection_bev_matrix() @ np.array([x, z, np.ones_like(u)])
    bev_image_points = np.round(bev_image_points).astype(np.int_)

    # consider points which are in bev grid
    bev_mask = (bev_image_points[0, :] >= 0) & (bev_image_points[0, :] < 200) & \
               (bev_image_points[1, :] >= 0) & (bev_image_points[1, :] < 200)

    bev_cls_map = np.zeros((200, 200))
    bev_cls_map[bev_image_points[1, :][bev_mask], bev_image_points[0, :][bev_mask]] = 1
    return bev_cls_map
    
def generate_BEV_projection(seg_map, depth_map, K, seg_cs=None):
    is_thresh = False
    
    if is_thresh and seg_cs is not None:
        depth_thresh = 50
        cs_thresh = 0.99
        seg_map[depth_map>depth_thresh] = 0
        seg_map[seg_cs<cs_thresh] = 0
    
    num_classes = 20
    bev_seg_map = np.zeros((num_classes, 200, 200))
    for cls_idx in range(num_classes):
        v, u = np.where(seg_map == cls_idx)
        z = depth_map[seg_map == cls_idx]
        bev_seg_map[cls_idx, :, :] = project_to_bev(u, z, K)

    return bev_seg_map

def show_bev_map(seg_id, depth, K, cs=None):
    bev_map_onehot = generate_BEV_projection(seg_id, depth, K, seg_cs=cs)
    bev_map_id = np.argmax(bev_map_onehot, axis=0)
    color_bev_map = np.zeros((bev_map_id.shape[0], bev_map_id.shape[1], 3), dtype=np.uint8)
    for label, color in enumerate(PALETTE_1):
        color_bev_map[bev_map_id == label, :] = color
        
    plt.imshow(color_bev_map)
    plt.show()      
    
def vis_func_1(rgb, depth, gt):
    fig = plt.figure(figsize=(14, 7))
    nrow = 1
    ncol = 3

    gs = gridspec.GridSpec(nrow, ncol,
                        wspace=0.0, hspace=-0.56, 
                        top=1.-0.5/(nrow+1), bottom=0.5/(nrow+1), 
                        left=0.5/(ncol+1), right=1-0.5/(ncol+1))

    ax1 = fig.add_subplot(gs[0,0])
    im1 = ax1.imshow(rgb)
    ax1.set_yticklabels([])
    ax1.set_xticklabels([])
    ax1.axis('off')

    ax2 = fig.add_subplot(gs[0,1])
    im2 = ax2.imshow(depth)
    ax2.set_yticklabels([])
    ax2.set_xticklabels([])
    ax2.axis('off')

    ax3 = fig.add_subplot(gs[0,2])
    im3 = ax3.imshow(gt)
    ax3.set_yticklabels([])
    ax3.set_xticklabels([])
    ax3.axis('off')

    plt.show() 
```
